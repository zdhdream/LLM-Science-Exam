```python
import re
import faiss
import numpy as np


wiki_text_data = ["===标题1===", "这是标题1的内容", "===标题2===", "这是标题2的内容"]

# 提取关键字（标题）
def extract_keyword(text):
    pattern = r"===(.*?)==="
    matches = re.findall(pattern, text)
    return matches

# 创建 Faiss 索引
def create_faiss_index(embeddings):
    index = faiss.IndexFlatL2(embeddings.shape[1])  # 使用L2距离度量
    index.add(embeddings)
    return index

# 示例嵌入向量，你需要替换成实际的嵌入向量
embeddings = np.random.rand(len(wiki_text_data), 100)

# 提取关键字并创建 Faiss 索引
keywords = [extract_keyword(text) for text in wiki_text_data]
keywords = sum(keywords, [])  # 将嵌套列表展平
keywords = np.array(keywords)  # 将关键字转换为NumPy数组
faiss_index = create_faiss_index(embeddings)

# 示例查询文本
query_text = "查询文本，这是一个示例 ===标题1==="
query_keywords = extract_keyword(query_text)
query_keywords = np.array(query_keywords)  # 将查询关键字转换为NumPy数组

# 在 Faiss 索引上执行标题搜索
k_title = 5  # 标题搜索结果的数量
query_embedding_title = np.random.rand(1, 100)  
D_title, I_title = faiss_index.search(query_embedding_title, k_title)

# 在 Faiss 索引上执行关键字搜索
k_keywords = 5  # 关键字搜索结果的数量
query_embedding_keywords = np.random.rand(1, 100)  
D_keywords, I_keywords = faiss_index.search(query_embedding_keywords, k_keywords)

# 合并标题搜索和关键字搜索的结果
merged_results = set(I_title[0]) | set(I_keywords[0])  # 合并两种搜索结果的索引

# 确保至少收集5个条目，最多收集10个条目
if len(merged_results) < 5:
    merged_results |= set(I_title[0][:5 - len(merged_results)])
if len(merged_results) > 10:
    merged_results = list(merged_results)[:10]

# 输出搜索结果
for index in merged_results:
    keyword = keywords[index]  # 获取关键字
    print(f"关键字: {keyword}")
```
